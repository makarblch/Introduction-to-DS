**Задание 7. Как зависят оценки от того, проходил ли студент курс для подготовки к сдаче экзамена (test preparation course)? Выведите на экран для каждого предмета в отдельности средний балл студентов, проходивших курс для подготовки к экзамену и не проходивших курс.**

```python
# Можно добавить бинарный столбец прохождения курса для подготовки (на основе обычного) и посчитать корреляцию
# Функция-конвертор
def fun(res):
  if res == 'completed':
    return 1
  elif res == 'none':
    return -1
  else:
    return 0


# Выведем значения корреляции для всех предметов
df['binary test preparation course'] = df['test preparation course'].apply(fun)
print("math corr:", df['binary test preparation course'].corr(df['math score']))
print("reading corr:", df['binary test preparation course'].corr(df['reading score']))
print("writing corr:", df['binary test preparation course'].corr(df['writing score']))

# А теперь выведем средние баллы
print("Math:")
print("With course: ", df[df['test preparation course'] == 'completed']["math score"].mean())
print("Without course: ", df[df['test preparation course'] == 'none']["math score"].mean())

print("Reading:")
print("With course: ", df[df['test preparation course'] == 'completed']["reading score"].mean())
print("Without course: ", df[df['test preparation course'] == 'none']["reading score"].mean())

print("Writing:")
print("With course: ", df[df['test preparation course'] == 'completed']["writing score"].mean())
print("Without course: ", df[df['test preparation course'] == 'none']["writing score"].mean())
```

***Видим, что средние баллы тех, кто посещал курсы для подготовки выше, чем баллы тех, кто не посещал их. Об этом может свидетельствовать и коэффициент корреляции (особенно в экзамене по письму - writing)***

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
%%time
eth = dict()
for index, element in df.iterrows():
  race = element['race/ethnicity']
  if race in eth:
    eth[race].append(element['reading score'])
  else:
    eth[race] = [element['reading score']]
eth = dict(sorted(eth.items()))
for race, value in eth.items():
  print(race, sum(value)/len(value))
```

```python
%%time
eth = dict()
for index, element in df.iterrows():
  race = element['education']
  if race in eth:
    eth[race].append(element['writing score'])
  else:
    eth[race] = [element['writing score']]
eth = dict(sorted(eth.items()))
for race, value in eth.items():
  print(race, sum(value)/len(value))
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def passed(grade):
  if grade < passmark:
    return 'F'
  return 'T'

# Math
df['Math_PassStatus'] = df['math score'].apply(passed)
print("Passed math test: ", len(df[df['Math_PassStatus'] == 'T']))
print("Failed math test: ", len(df[df['Math_PassStatus'] == 'F']))
# Reading
df['Reading_PassStatus'] = df['reading score'].apply(passed)
print("Passed reading test: ", len(df[df['Reading_PassStatus'] == 'T']))
print("Failed reading test: ", len(df[df['Reading_PassStatus'] == 'F']))
# Writing
df['Writing_PassStatus'] = df['writing score'].apply(passed)
print("Passed writing test: ", len(df[df['Writing_PassStatus'] == 'T']))
print("Failed writing test: ", len(df[df['Writing_PassStatus'] == 'F']))
df.head(5)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
      return 'A'
    elif 80 < average_mark <= 90:
      return 'B'
    elif 70 < average_mark <= 80:
      return 'C'
    elif 60 < average_mark <= 70:
      return 'D'
    elif 50 < average_mark <= 60:
      return 'E'
    return 'F'

df['Grade'] = ((df['math score'] + df['reading score'] + df['writing score']) / 3).apply(GetGrade)
df_grades = df['Grade'].value_counts()
print(df_grades)
df.head(5)
```
